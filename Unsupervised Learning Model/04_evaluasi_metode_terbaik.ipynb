```python
# 04_evaluasi_metode_terbaik.ipynb

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load preprocessed data
data = pd.read_csv('preprocessed_data.csv')

# Pilih metode terbaik: K-Means (misalnya, k=4)
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(data)
labels = kmeans.labels_

# PCA untuk visualisasi
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

# Visualisasi hasil clustering
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
plt.colorbar(label='Cluster')
plt.title('Visualisasi Clustering dengan PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()

# Simpan hasil clustering
pd.DataFrame({'Cluster': labels}).to_csv('clustering_results.csv', index=False)
print("Evaluasi selesai! Hasil clustering disimpan ke 'clustering_results.csv'.")
```
